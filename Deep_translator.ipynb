```python
from deep_translator import GoogleTranslator

languages = {
    "ta": "Tamil",
    "hi": "Hindi",
    "fr": "French",
    "es": "Spanish",
    "de": "German"
}

print("===== LANGUAGE TRANSLATION TOOL =====")

while True:
    text = input("\nEnter text to translate: ").strip()

    if not text:
        print("Please enter some text.")
        continue

    print("\nAvailable Languages:")
    for code, name in languages.items():
        print(f"{code} - {name}")

    target = input("\nEnter target language code: ").lower().strip()

    if target not in languages:
        print("Invalid language code!")
        continue

    try:
        translated = GoogleTranslator(source="auto", target=target).translate(text)

        print("\nTranslated Text:")
        print(translated)

    except Exception as e:
        print("Error:", e)

    again = input("\nTranslate another text? (yes/no): ").lower().strip()

    if again != "yes":
        print("Thank you for using the translator!")
        break
```
